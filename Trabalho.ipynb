```r
n = length(ordenados)
assimetria = (sum((ordenados - mean(ordenados))^3)/n)/(sum((ordenados - mean(ordenados))^2)/n)^(3/2)
u = ordenados
curtose <- n * sum((u - mean(u))^4)/(sum((u - mean(u))^2)^2)
q1 = ordenados[10]+0.75*(ordenados[11]-ordenados[10])
q3 = quantile(ordenados)[4]
#boxplot(viscosidade, main = "Histograma dos dados",xlab="Dados",)
bp1 = c(min(ordenados),q1,median(ordenados),q3,
        max(viscosidade),
        assimetria,
        curtose,
        max(viscosidade)-min(viscosidade))
boxplot(bp1,
        col = "orange",
        border = "brown",
horizontal = TRUE

)
box()
```

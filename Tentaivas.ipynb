### 2.1 HISTOGRAMA 

```r
ordenados = c(42.8, 42.9, 43.0, 43.0, 43.0, 43.0, 43.1, 43.1, 43.2, 43.2, 43.2, 43.3, 43.5, 43.5, 43.6, 43.6, 47.2, 47.5, 47.9, 47.9, 47.9, 48.0, 48.0, 48.0, 48.0, 48.0, 48.1, 48.1, 48.1, 48.3, 48.3, 48.3, 48.4, 48.5, 48.6, 48.6, 48.6, 48.6, 48.8, 48.9)
AT = max(ordenados)-min(ordenados)
hist(viscosidade,breaks=7, main = "Histograma dos dados", xlab = "Dados", ylab="Frequência",col="lightblue")
box()
```

```r
n = length(ordenados)
assimetria = (sum((ordenados - mean(ordenados))^3)/n)/(sum((ordenados - mean(ordenados))^2)/n)^(3/2)
u = ordenados
curtose <- n * sum((u - mean(u))^4)/(sum((u - mean(u))^2)^2)
q1 = ordenados[10]+0.75*(ordenados[11]-ordenados[10])
q3 = quantile(ordenados)[4]
#boxplot(viscosidade, main = "Histograma dos dados",xlab="Dados",)
bp1 = c(min(ordenados),q1,median(ordenados),q3,
        max(viscosidade),
        assimetria,
        curtose,
        max(viscosidade)-min(viscosidade))
boxplot(bp1,
        col = "orange",
        border = "brown",
horizontal = TRUE

)
box()
```
